```python
endoint = f"{url}/login"

payload = json.dumps({
  "username": "testuser",
  "email": "test@example.com",
  "password": "testpassword"
})
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("POST", endoint, headers=headers, data=payload)

print(response.text)

login_response = json.loads(response.text)
```

```python
import requests
import json

endoint = f"{url}/generate_random_array"

payload = json.dumps({
  "sentence": "Hello World!!"
})
headers = {
  'Authorization': f"Bearer {login_response['access_token']}",
  'Content-Type': 'application/json'
}

response = requests.request("POST", endoint, headers=headers, data=payload)

print(response.text)
```
